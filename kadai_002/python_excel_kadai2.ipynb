```python
#　pandasをインポートし、pdという別名でpandasを扱えるようにする
import pandas as pd
import openpyxl
# openpyxl.stylesモジュールのFontクラスをインポート
from openpyxl.styles import Font

#　データフレームを定義する
df = pd.DataFrame(data = {"日付": ["2023-05-17", "2023-05-18",
    "2023-05-19", "2023-05-20", "2023-05-21"],
    "社員名": ["山田", "佐藤", "鈴木", "田中","高橋"],
    "売上": [100, 200, 150, 300, 250],
    "部門": ["メーカー", "代理店", "メーカー", "商社", "代理店"]})

#　「平均売上」列を作成し、全体の売上の平均を算出する
df["平均売上"] = df["売上"].mean()
#　「平均売上」を変数「average_sales」に代入
average_sales = df["売上"].mean()

#　関数「performance」を定義
def performance(sales):
    #　「平均売上」はfloat64であるため、「売上」をfloat64に変換
    df["売上（小数点型）"] = df["売上"] .astype("float64")
    #　変数「average_sales」の値と引数「sales」の値を比較し、「業績ランク」の条件分岐を行う
    rank =""
    if sales >= average_sales + 50:
        rank = "A"
    elif sales >= average_sales:
        rank = "B"
    else: rank = "C"
    return rank

#　「業績ランク」列を作成
#　関数「performance」から「売上」を引数「sales」に与えて「業績ランク」に入力
df["業績ランク"] = df["売上"].apply(performance)

#　「売上（小数点型）」の列を削除
df.drop("売上（小数点型）", axis=1, inplace=True)

#　Excelファイルにデータフレームの内容を書き込み、保存
writer = pd.ExcelWriter("業績.xlsx")
df.to_excel(writer, sheet_name="Sheet1", index=False)
writer.close()

# Excelファイルを読み込み
wb = openpyxl.load_workbook("業績.xlsx")
ws = wb["Sheet1"]

# 設定したいフォントの定義(フォント名、サイズ、太字、色)
new_font = Font(name="Yu Gothic UI",  size=11, bold=True, color="FF000000")

# 全行・全列のセルにフォントを適用
for row in ws.iter_rows():
    for cell in row:
        cell.font = new_font

#　各列の幅を調整
ws.column_dimensions["A"].width = 19
ws.column_dimensions["B"].width = 12
ws.column_dimensions["C"].width = 9
ws.column_dimensions["D"].width = 13
ws.column_dimensions["E"].width = 15
ws.column_dimensions["F"].width = 16

wb.save("業績.xlsx")
```
