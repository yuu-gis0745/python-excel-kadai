```python
import openpyxl
# timeモジュールをインポート
import time

# Workbookクラスのインスタンス化
workbook = openpyxl.Workbook()
# アクティブなシートを変数「ws」に格納する
ws = workbook.active

# セルに値を書き込む
ws["B2"] = "請求書"
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = time.strftime("%Y/%m/%d",time.localtime())
# 系列データのタイトル設定
ws.cell(row=10, column=2, value="商品名")
ws.cell(row=10, column=3, value="数量")
ws.cell(row=10, column=4, value="単価")
ws.cell(row=10, column=5, value="金額")
ws.cell(row=10, column=2, value="商品名")

# 商品名
product = ["商品A","商品B"]
# 数量
count =[2,1]
# 単価
price = [10000, 15000]

# 小計を計算するための変数「small_price」を作成
small_price=[]

for i, name in enumerate(product):
    row=i+11
    # 商品名（B列）
    ws.cell(row=row, column=2, value=name)
    # 数量（C列）
    ws.cell(row=row, column=3, value=count[i])
    # 単価（D列）
    ws.cell(row=row, column=4, value=price[i])
    # 金額（E列）
    product_price = count[i] * price[i]
    ws.cell(row=row, column=5, value=product_price)
    # 金額を計算するごとに、小計算出用の変数に追加
    small_price.append(product_price)

# 小計を算出
small_all_price = sum(small_price)

# 小計の値を金額の下に挿入
ws.cell(row=11+len(product), column=5, value=small_all_price)

# 集計結果のタイトルを設定する
ws.cell(row=11+len(product)+2, column=2, value="小計")
ws.cell(row=11+len(product)+3, column=2, value="消費税")
ws.cell(row=11+len(product)+4, column=2, value="合計")

# 1行空けて小計の値をセルに挿入
ws.cell(row=11+len(product)+2, column=5, value=small_all_price)

# 消費税の計算
def calculate_tax(small_all_price, tax_rate=0.1):
    tax = int(small_all_price * tax_rate)
    return tax
# 消費税の額をセルに挿入
ws.cell(row=11+len(product)+3, column=5, value=calculate_tax(small_all_price))

# 消費税を含めた合計金額を算出
total_price = small_all_price + calculate_tax(small_all_price)

# 合計金額をセルに挿入
ws.cell(row=11+len(product)+4, column=5, value=total_price)

# 仮のExcelファイルを保存
workbook.save("test.xlsx")
```

```python
import openpyxl
# openpyxl.stylesモジュールのFontクラスをインポート
from openpyxl.styles import Font

# Excelファイルを読み込み
workbook = openpyxl.load_workbook("test.xlsx")
sheet = workbook['Sheet']

# 設定したいフォントの定義(フォント名、サイズ、太字、色)
new_font = Font(name="Yu Gothic UI",  size=10, bold=True, color="FF000000")


# 全行・全列のセルにフォントを適用
for row in sheet.iter_rows():
    for cell in row:
        cell.font = new_font

# 今日の日付でExcelファイルを保存
today = time.strftime("%Y年%m月%d日", time.localtime())
workbook.save(f"請求書_{today}.xlsx")
```
